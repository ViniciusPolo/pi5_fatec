```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pandas.plotting import scatter_matrix

preferences = pd.read_csv('base_food_preference/Food_Preference.csv')

preferences.head()
```

```python
preferences = preferences.drop(columns=['Timestamp','Participant_ID'])

preferences.head()
```

```python
preferences_duplicates = preferences [ preferences.duplicated(subset=['Gender', 'Nationality', 'Age', 'Food', 'Juice', 'Dessert'], keep=False)]

if len(preferences_duplicates) > 0:
  print('\nAmostras redudantes ou inconsistentes:\n')
  print(preferences_duplicates)
else: print('\nNÃ£o existem amostras redudantes ou inconsistentes')
```

```python
def delDuplicatas(preferences):
   '''
   Para cada grupo de amostras duplicadas, mantÃ©m uma e apaga as demais
   '''
  
   # remove as amostras duplicadas, mantendo apenas a primeira ocorrencia
   preferences = preferences.drop_duplicates(keep = 'first')   
 
   return preferences
 
preferences = delDuplicatas(preferences)
 
display(preferences)
```
